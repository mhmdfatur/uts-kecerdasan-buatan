```python
employee_detail_list = [
    {'full_name': 'John Doe', 'company': 'Company 1'},
    {'full_name': 'Tom Smith', 'company': 'Company 2'},
    {'full_name': 'Andrew Sebastian', 'company': 'Company 3'},
    {'full_name': 'Ying Han Tan', 'company': 'Company 1'},
    {'full_name': 'Kenneth Ng', 'company': 'Company 3'},
    {'full_name': 'Rubby Lee', 'company': 'Company 2'},
    {'full_name': 'Robert White', 'company': 'Company 1'},
]

company_detail_list = [
    {'name': 'Company 1', 'country': 'United States'},
    {'name': 'Company 2', 'country': 'United Kingdom'},
    {'name': 'Company 3', 'country': 'United Kingdom'},
]

def employees_with_company_country():
    result = []
    for employee in employee_detail_list:
        company_name = employee['company']
        company_country = next(
            (company['country'] for company in company_detail_list if company['name'] == company_name), None
        )
        if company_country:
            result.append({
                'full_name': employee['full_name'],
                'company': company_name,
                'country': company_country
            })
    return result

employees = employees_with_company_country()
print("[")
for item in employees:
    print("    {")
    print(f'        "full_name": "{item["full_name"]}",')
    print(f'        "company": "{item["company"]}",')
    print(f'        "country": "{item["country"]}"')
    print("    },")
print("]")
```

```python
def get_companies_with_employee_nationality():
    employees = [
        {'company': 'Company 1', 'nationality': 'Australia'},
        {'company': 'Company 1', 'nationality': 'Australia'},
        {'company': 'Company 1', 'nationality': 'Hong Kong'},
        {'company': 'Company 2', 'nationality': 'Hong Kong'},
        {'company': 'Company 2', 'nationality': 'United States'},
        {'company': 'Company 3', 'nationality': 'United States'},
        {'company': 'Company 3', 'nationality': 'United States'},
    ]
    
    company_nationality_count = {}
    
    for emp in employees:
        company = emp['company']
        nationality = emp['nationality']
        company_nationality_count.setdefault(company, {}).setdefault(nationality, 0)
        company_nationality_count[company][nationality] += 1
    
    # Mengurutkan dan menampilkan hasil
    for company, nationalities in sorted(company_nationality_count.items()):
        print(f"Company: {company}")
        for nationality, count in nationalities.items():
            print(f"  {nationality}: {count}")
        print()  # Baris kosong untuk pemisah antar perusahaan

# Memanggil fungsi
get_companies_with_employee_nationality()    
```
